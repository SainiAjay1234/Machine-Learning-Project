```python
#Under-Sampling
#Build a sample dataset containing similar distribution of normal transaction and Fradulent
#Number of Fradulent transaction--> 492 
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

Splitting the data into features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1) # almost full dataset except class
Y = new_dataset['Class'] #class containing dataset
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

#Logistic Regression


```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's.

```python
#Model Evaluation based on accuracy score
```

Model Evaluation

Accuracy Score:


```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#Accuracy score  on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score of test data :", test_data_accuracy)
```
